Estos son conjuntos de datos que consisten en 3 tipos diferentes de iris (Setosa, Versicolor y Virginica)

| Nombre de Varibele | Rol | Tipo | Descripsción | Unidades | Missing|
|---|---|---|---|---|---|

|Sepal width | Feature | Continua | Anchura del Sepalo | cm | no |
|Petal length | Feature | Continua | Longuitud del Petalo | cm | no |
|Petal width | Feature | Continua | Anchura del Petalo | cm | no |
|Class | Target | Categoria | Clase de iris: Setosa, Iris Versicolor o Iris Virginica | 0,1,2 | no |

# Preparación de Datos

### Librearías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats 
from sklearn.datasets import load_iris
```

```python
data = load_iris()
data.keys()
```

```python
variables = pd.DataFrame(data['data'], columns=data['feature_names'])
variables
```

```python
variables['clase'] = data['target']
variables
```

```python
data['target_names']
```

```python
# esta es la forma cuando tenemos muchos elementos.
#dict(zip(range(len(data['target_names'])), data['target_names']))
```

```python
variables['clase'].replace({0:'setosa', 1:'versicolor', 2:'virginica'}, inplace=True)
variables.to_parquet('Iris.parquert', index=False) #guardar dataframe
```

```python
df = pd.read_parquet('Iris.parquert')
df
```

## Procesado

**Feature Engineer**

```python
df['area sepal (cm2)'] = df['sepal length (cm)'] * df['sepal width (cm)']
df['area petal (cm2)'] = df['petal length (cm)'] * df['petal width (cm)']
```
