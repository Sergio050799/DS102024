### Paso 3. Asígnalo a una variable llamada usuarios.

```python
usuario = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep= '|')
usuario
```

### Paso 4. Descubre cuál es la edad media por ocupación

```python
usuario.groupby('occupation')[['age']].mean()
```

### Paso 5. Descubra la proporción de Hombres por ocupación y ordénela de mayor a menor

#### Forma 1

```python
usuario[usuario['gender'] == 'M'].groupby('occupation').size() / usuario.groupby('occupation').size().sort_values(ascending=False) * 100
```

```python
usuario.groupby(['occupation'])['gender'].value_counts(normalize=True).xs('M',level='gender')
```

#### Forma 2

```python
usuario.groupby(['occupation', 'gender'], as_index=False)['age'].count()
```

#### Forma 3

### Paso 6. Para cada ocupación, calcule las edades mínima y máxima

```python
usuario.groupby('occupation')['age'].agg(['min', 'max'])
```

### Paso 7. Para cada combinación de ocupación y sexo, calcule la edad media

```python
usuario.groupby(['occupation', 'gender'])['age'].mean()
```

### Paso 8.  Para cada ocupación, presente el porcentaje de mujeres y hombres

```python
usuario.groupby('occupation')['gender'].value_counts(normalize=True).unstack().fillna(0) * 100
```
